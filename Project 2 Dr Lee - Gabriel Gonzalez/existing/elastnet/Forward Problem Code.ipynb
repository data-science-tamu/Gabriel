```python
import deepxde as dde
import numpy as np
import torch
from sklearn import preprocessing
import time
import matplotlib.pyplot as plt

# Load and preprocess data
x_disp = np.loadtxt('data_incompressible/m_rose_nu_05/disp_coord')
y_disp = np.loadtxt('data_incompressible/m_rose_nu_05/disp_data')   # ux, uy displacement data
x_strain = np.loadtxt('data_incompressible/m_rose_nu_05/strain_coord')
y_strain = np.loadtxt('data_incompressible/m_rose_nu_05/strain_data')
m_data = np.loadtxt('data_incompressible/m_rose_nu_05/m_data')       # Known Young's modulus values
nu_data = np.loadtxt('data_incompressible/m_rose_nu_05/nu_data')     # Known Poisson's ratio values

# Check for NaNs in input data
print("NaNs in x_disp:", np.isnan(x_disp).any())
print("NaNs in y_disp:", np.isnan(y_disp).any())
print("NaNs in x_strain:", np.isnan(x_strain).any())
print("NaNs in y_strain:", np.isnan(y_strain).any())
print("NaNs in m_data:", np.isnan(m_data).any())
print("NaNs in nu_data:", np.isnan(nu_data).any())

# Normalize coordinates (optional)
ss_x = preprocessing.StandardScaler()
x_disp_norm = ss_x.fit_transform(x_disp)
x_strain_norm = ss_x.transform(x_strain)  # Use the same scaler to transform x_strain

# Define the geometry based on normalized data
x_min, y_min = np.min(x_disp_norm, axis=0)
x_max, y_max = np.max(x_disp_norm, axis=0)
geom = dde.geometry.Rectangle([x_min, y_min], [x_max, y_max])

def E_true(x):
    # Ensure x is converted to numpy array only if it's a tensor
    if isinstance(x, torch.Tensor):
        x = x.detach().cpu().numpy()  # Convert to numpy if x is a tensor
    return 1.0 + np.sin(x[:, 0:1])  # E_true varies with sin(x) over the x-coordinates


# Define the force function with sin(x) applied as a body force
def force_function(x):
    return torch.sin(x[:, 0:1])  # Force varies with x-coordinate

# Define the PDE residuals for the forward problem with an external force
def pde_forward(x, u):
    ux = u[:, 0:1]  # Predicted displacement u_x
    uy = u[:, 1:2]  # Predicted displacement u_y
    E = torch.tensor(E_true(x), dtype=torch.float32)  # Convert to tensor if necessary

    
    # Poisson's ratio, assumed constant here
    nu = 0.3

    # Compute derivatives
    ux_x = dde.grad.jacobian(ux, x, i=0, j=0)
    ux_y = dde.grad.jacobian(ux, x, i=0, j=1)
    uy_x = dde.grad.jacobian(uy, x, i=0, j=0)
    uy_y = dde.grad.jacobian(uy, x, i=0, j=1)

    # Stress tensor relations
    sigma_xx = E * (ux_x + nu * uy_y)
    sigma_yy = E * (uy_y + nu * ux_x)
    tau_xy = E * (ux_y + uy_x) / (2 * (1 + nu))

    # External force (sin(x)) in x and y directions
    fx = force_function(x)  # Force in the x-direction
    fy = force_function(x)  # Assuming the same force in the y-direction

    # Residuals for equilibrium equations with external force
    res_x = dde.grad.jacobian(sigma_xx, x, i=0, j=0) + dde.grad.jacobian(tau_xy, x, i=0, j=1) - fx
    res_y = dde.grad.jacobian(sigma_yy, x, i=0, j=1) + dde.grad.jacobian(tau_xy, x, i=0, j=0) - fy

    return [res_x, res_y]

# Define boundary conditions using observed displacement data as Dirichlet BCs
bc_ux = dde.icbc.PointSetBC(x_disp_norm, y_disp[:, 0:1], component=0)  # u_x from y_disp
bc_uy = dde.icbc.PointSetBC(x_disp_norm, y_disp[:, 1:2], component=1)  # u_y from y_disp

# Data setup for forward problem
data = dde.data.PDE(
    geom,
    pde_forward,
    [bc_ux, bc_uy],  # Use displacement data as boundary conditions
    num_domain=2000,
    num_boundary=100,
)

# Neural network architecture (predicting ux and uy)
layer_size = [2] + [64] * 2 + [2]  # 2 input dims, 2 hidden layers, 2 outputs (ux, uy)
activation = "relu"
initializer = "Glorot normal"
net = dde.nn.FNN(layer_size, activation, initializer)

# Output transformation to ensure E_pred > 0 using softplus
def modify_output(inputs, outputs):
    E_pred = torch.nn.functional.softplus(outputs[:, 0:1])  # Ensure E_pred > 0
    return torch.cat([E_pred, outputs[:, 1:]], dim=1)  # Pass through ux and uy unmodified

# Modify the network to apply this output transformation
net.apply_output_transform(modify_output)

# Model definition and compilation
model = dde.Model(data, net)
model.compile("adam", lr=0.0005)

# Start the timer
start_time = time.time()

# Train the model
losshistory, train_state = model.train(iterations=100, display_every=1)

# Save elapsed time
print("--- %s Elapsed time ---" % (time.time() - start_time))

# Prediction for displacements using x_strain_norm for evaluation
ux_pred = model.predict(x_strain_norm)[:, 0]  # Predicted u_x
uy_pred = model.predict(x_strain_norm)[:, 1]  # Predicted u_y

# Save predicted displacements
np.savetxt('ux_pred_forward', ux_pred)
np.savetxt('uy_pred_forward', uy_pred)

# Visualization of the displacements if needed
plt.figure(figsize=(10, 8))
plt.scatter(x_strain_norm[:, 0], x_strain_norm[:, 1], c=ux_pred, cmap="viridis", marker='o')
plt.colorbar(label="Predicted Displacement (ux)")
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.title("Predicted Displacement Field (ux) for Forward Problem")
plt.show()

plt.figure(figsize=(10, 8))
plt.scatter(x_strain_norm[:, 0], x_strain_norm[:, 1], c=uy_pred, cmap="viridis", marker='o')
plt.colorbar(label="Predicted Displacement (uy)")
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.title("Predicted Displacement Field (uy) for Forward Problem")
plt.show()
```
